```python
import numpy as np
import torch

import torch.nn as nn
import torch.optim as optim
import torch.nn.init as init
import matplotlib.pyplot as plt

# x 데이터: 150 ~ 200 범위에서 랜덤 생성(100개)
# x 데이터: 입력값이자 키 데이터
x = 150 + (torch.rand(100, 1) * 50)

# target 데이터: 40 ~ 100 범위에서 랜덤 생성(100개)
# target 데이터: 목푯값이자 몸무게 데이터
target = 40 + (torch.rand(100, 1) * 60)

# # 정규화
# x_min, x_max = x.min(), x.max()
# target_min, target_max = target.min(), target.max()

# x = (x - x_min) / (x_max - x_min)
# target = (target - target_min) / (target_max - target_min)

print(x.shape)
print(target.shape)
```

```python
num_epoch = 500
loss_arr = []

# 선형 모델 정의
model = nn.Linear(1, 1)  # 입력값 1개(키), 출력값 1개(몸무게)

# 손실 함수
loss_func = nn.MSELoss()

# 최적화
optimizer = optim.SGD(model.parameters(), lr=0.00001)

for i in range(num_epoch):
  optimizer.zero_grad()
  output = model(x)

  loss = loss_func(output, target)
  print("Current loss: ", loss.item())
  loss.backward()
  optimizer.step()

  if i % 10 == 0:
    plt.scatter(x.detach().numpy(), output.detach().numpy())
    plt.axis([140, 210, 30, 110])
    plt.show()
    print(loss.data)

  loss_arr.append(loss.detach().numpy())

# 최종 모델 파라미터 출력
print("\n최종 가중치와 절편:")
print(list(model.parameters()))
```
